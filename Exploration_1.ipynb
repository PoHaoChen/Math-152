# **The Collatz Conjecture**

The Collatz conjecture is, for any positive integer n:
1.   If n is even, replace it by n/2.
2.   If n is odd, replace it by 3n+1.

By keep replacing n with this rule, we will finally get in to the cycle 4, 2, 1 at the end. This conjecture is not proved to be true yet (2/9/2022).

There are three parts in this exploration. The first part will be test the conjecture up to one million. The second part will be about inputing negative number in the conjecture. The third part will be exploring when "n" will be lower than the original "n" during the process.





**Part One:**

In the following code, input "n" will be process until n=1 by the rule of the Collatz conjecture. The loop will be running at least three times. A print method can also be added into the program to print the number of step taken to reach the result.

```python
def the_Collatz_conjecture(n):
  ori_n=n
  result=[]
  step_count=0
  inf_stoper=0
  drop_timer=0
  while abs(n)!=1 or len(result)<3:
    result.append(n)
    step_count+=1 #this is for counting the number of steps
    if abs(n)%2==0:
      n=int(n/2)
    else:
      n=3*n+1
    #if inf_stoper>10000: #this is for testing negative numbers
      #break
    #inf_stoper+=1
    #if ori_n>n:
      #drop_timer=step_count
      #break
  if n>0:
    result.append(1)
  else:
    result.append(-1)
  return result
```

The following code will check if the list returned from previous program end up with 4, 2, 1 cycle. Notice the program takes about a minute to check until one million. In the argument of the while loop, the number of x can be modified to change to number it checks.

```python
# Check until 1 million (will take a minute)
Checker=True
Collatz_seq=[]
x=1
while Checker and x<=1000000:
  Collatz_seq=the_Collatz_conjecture(x)
  Checker=(Collatz_seq[len(Collatz_seq)-3]==4) and (Collatz_seq[len(Collatz_seq)-2]==2) and (Collatz_seq[len(Collatz_seq)-1]==1)
  if not Checker:
    print("{} does not follow the conjecture.".format(x))
  x+=1


```

Based on the program, the Collatz conjecture is true until one million.
